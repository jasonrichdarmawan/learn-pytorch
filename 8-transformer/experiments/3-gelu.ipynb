```python
import math
import numpy as np

import matplotlib.pyplot as plt

def gelu(x):
  """
  Compute the Gaussian Error Linear Unit (GELU) activation.
  This version uses the tanh approximation.
  """
  return 0.5 * x * (1 + np.tanh(math.sqrt(2 / math.pi) * (x + 0.044715 * x**3)))

def relu(x):
  return np.maximum(0, x)

# Generate values to plot the GELU activation function
x = np.linspace(-5, 5, 1000)
y = gelu(x)
y_relu = relu(x)

plt.figure(figsize=(8,5))
plt.plot(x, gelu(x), label='GELU(x)')
plt.plot(x, y_relu, label='ReLU(x)', linestyle='--')
plt.title('GELU vs ReLU Activation Functions')
plt.xlabel('Input')
plt.ylabel('Activation')
plt.grid(True)
plt.legend()
plt.show()
```
