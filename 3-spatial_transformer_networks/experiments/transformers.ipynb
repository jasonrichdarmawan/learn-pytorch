```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import torchvision.transforms.functional as F

# Create a sample image (white background with a blue rectangle)
img = Image.new('RGB', (256, 256), 'white')
draw = ImageDraw.Draw(img)
draw.rectangle([50, 50, 100, 100], outline='blue', width=5)

# Create a corresponding mask (black background with a white rectangle)
mask = Image.new('L', (256, 256), 0)  # 'L' mode for grayscale mask
draw_mask = ImageDraw.Draw(mask)
draw_mask.rectangle([50, 50, 100, 100], fill=255)

# Transformation: rotate by 45 degrees
angle = 45
img_transformed = F.rotate(img, angle)
mask_transformed = F.rotate(mask, angle)

# Visualize original and transformed images and masks
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].imshow(img)
axs[0, 0].set_title("Original Image")
axs[0, 1].imshow(mask, cmap='gray')
axs[0, 1].set_title("Original Mask")
axs[1, 0].imshow(img_transformed)
axs[1, 0].set_title("Transformed Image")
axs[1, 1].imshow(mask_transformed, cmap='gray')
axs[1, 1].set_title("Transformed Mask")

# Remove axis for clarity
for ax in axs.flat:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
